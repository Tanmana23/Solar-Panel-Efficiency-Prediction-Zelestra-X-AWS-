```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# First, analyze the categorical columns to understand their cardinality
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

print("Categorical Column Analysis:")
for col in categorical_cols:
    unique_count = df[col].nunique()
    print(f"{col}: {unique_count} unique values")

# Separate low and high cardinality columns
low_cardinality_cols = [col for col in categorical_cols if df[col].nunique() <= 20]
high_cardinality_cols = [col for col in categorical_cols if df[col].nunique() > 20]

print(f"\nLow cardinality columns (â‰¤20 categories): {low_cardinality_cols}")
print(f"High cardinality columns (>20 categories): {high_cardinality_cols}")

# Plot low cardinality columns directly
if low_cardinality_cols:
    fig, axes = plt.subplots(1, len(low_cardinality_cols), figsize=(6*len(low_cardinality_cols), 6))
    if len(low_cardinality_cols) == 1:
        axes = [axes]
    
    for i, col in enumerate(low_cardinality_cols):
        sns.boxplot(data=df, x=col, y='efficiency', ax=axes[i])
        axes[i].set_title(f'Efficiency by {col}')
        axes[i].tick_params(axis='x', rotation=45)
        
        # Add statistical annotation
        efficiency_by_cat = df.groupby(col)['efficiency'].agg(['mean', 'std', 'count'])
        print(f"\n{col} Statistics:")
        print(efficiency_by_cat)
    
    plt.tight_layout()
    plt.show()

# Handle high cardinality columns with statistical analysis
for col in high_cardinality_cols:
    print(f"\nAnalysis for {col} (High Cardinality)")
    
    # Statistical summary by category
    cat_stats = df.groupby(col)['efficiency'].agg(['mean', 'std', 'count']).sort_values('mean', ascending=False)
    
    # Show top and bottom performers
    print(f"Top 10 {col} by efficiency:")
    print(cat_stats.head(10))
    print(f"\nBottom 10 {col} by efficiency:")
    print(cat_stats.tail(10))
    
    # Plot only top 15 categories by frequency (most data points)
    top_categories = df[col].value_counts().head(15).index
    df_subset = df[df[col].isin(top_categories)]
    
    plt.figure(figsize=(14, 6))
    sns.boxplot(data=df_subset, x=col, y='efficiency')
    plt.title(f'Efficiency by Top 15 Most Frequent {col}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    
    # Calculate variance between categories (key insight for predictive power)
    overall_var = df['efficiency'].var()
    between_group_var = cat_stats['mean'].var()
    predictive_power = between_group_var / overall_var
    
    print(f"Predictive Power Score: {predictive_power:.4f}")

# Summary analysis for feature selection
print("FEATURE SELECTION INSIGHTS:")

for col in categorical_cols:
    if col in low_cardinality_cols:
        # ANOVA F-statistic for low cardinality
        from scipy import stats
        groups = [group['efficiency'].values for name, group in df.groupby(col)]
        f_stat, p_value = stats.f_oneway(*groups)
        print(f"{col}: F-statistic = {f_stat:.4f}, p-value = {p_value:.4f}")
        
    else:
        # Variance ratio for high cardinality
        cat_stats = df.groupby(col)['efficiency'].agg(['mean', 'count'])
        # Weight by sample size
        weighted_var = np.average((cat_stats['mean'] - df['efficiency'].mean())**2, 
                                weights=cat_stats['count'])
        predictive_score = weighted_var / df['efficiency'].var()
        print(f"{col}: Weighted Predictive Score = {predictive_score:.4f}")
```

```python
import pandas as pd
import numpy as np

df_cleaned = train_df.copy()

# Convert humidity, wind_speed, pressure to numeric (coerce errors to NaN)
numeric_cols_to_fix = ['humidity', 'wind_speed', 'pressure']
for col in numeric_cols_to_fix:
    before_conversion = df_cleaned[col].dtype
    # Count problematic values before conversion
    problematic_count = df_cleaned[col].astype(str).str.contains('unknown|badval', case=False, na=False).sum()
    
    # Convert to numeric, coercing errors to NaN
    df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')
    
    print(f"   {col}: {before_conversion} -> {df_cleaned[col].dtype}")
    print(f"      Converted {problematic_count} problematic values to NaN")

print()

# Fix negative irradiance (physically impossible)
negative_irradiance_count = (df_cleaned['irradiance'] < 0).sum()
df_cleaned.loc[df_cleaned['irradiance'] < 0, 'irradiance'] = 0
print(f"   Fixed {negative_irradiance_count} negative irradiance values -> 0")

# Fix cloud coverage > 100% (impossible percentage)
high_cloud_count = (df_cleaned['cloud_coverage'] > 100).sum()
df_cleaned.loc[df_cleaned['cloud_coverage'] > 100, 'cloud_coverage'] = 100
print(f"   Capped {high_cloud_count} cloud coverage values > 100% -> 100%")

# Fix extreme temperatures (likely sensor errors)
extreme_temp_count = (df_cleaned['temperature'] > 60).sum()
df_cleaned.loc[df_cleaned['temperature'] > 60, 'temperature'] = np.nan
print(f"   Set {extreme_temp_count} extreme temperatures > 60C -> NaN (for imputation)")

print()

# Handle error_code missing values (likely means "No Error")
error_code_missing = df_cleaned['error_code'].isnull().sum()
df_cleaned['error_code'] = df_cleaned['error_code'].fillna('No_Error')
print(f"   Filled {error_code_missing} missing error_code values -> 'No_Error'")

# Handle installation_type missing values with mode
installation_mode = df_cleaned['installation_type'].mode()[0]
installation_missing = df_cleaned['installation_type'].isnull().sum()
df_cleaned['installation_type'] = df_cleaned['installation_type'].fillna(installation_mode)
print(f"   Filled {installation_missing} missing installation_type values -> '{installation_mode}'")

print()

print("4. DATA TYPE OPTIMIZATION:")

# Convert categorical columns to category dtype for memory efficiency
categorical_cols = ['string_id', 'error_code', 'installation_type']
for col in categorical_cols:
    df_cleaned[col] = df_cleaned[col].astype('category')
    print(f"   {col}: object -> category")

print()

print("5. VALIDATION OF FIXES:")
# Validate fixes
print("   Negative irradiance remaining:", (df_cleaned['irradiance'] < 0).sum())
print("   Cloud coverage > 100% remaining:", (df_cleaned['cloud_coverage'] > 100).sum())
print("   Temperature > 60C remaining:", (df_cleaned['temperature'] > 60).sum())
print("   Error_code missing remaining:", df_cleaned['error_code'].isnull().sum())
print("   Installation_type missing remaining:", df_cleaned['installation_type'].isnull().sum())

print()

missing_summary = df_cleaned.isnull().sum().sort_values(ascending=False)
missing_summary = missing_summary[missing_summary > 0]

for col, count in missing_summary.items():
    percentage = (count / len(df_cleaned)) * 100
    print(f"   {col}: {count} ({percentage:.1f}%)")

print()

print(df_cleaned.dtypes)

```

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer, SimpleImputer

# Current missing data status
print("1. CURRENT MISSING DATA STATUS:")
missing_summary = df_cleaned.isnull().sum().sort_values(ascending=False)
missing_summary = missing_summary[missing_summary > 0]
for col, count in missing_summary.items():
    percentage = (count / len(df_cleaned)) * 100
    print(f"   {col}: {count} ({percentage:.1f}%)")
print()

print("2. FAST STRATEGIC IMPUTATION:")
# Create a copy for imputation
df_imputed = df_cleaned.copy()

# Strategy 1: Physics-based imputation for electrical features (FAST)
print("   Strategy 1: Physics-based imputation for electrical features")

# Calculate ratios only once for efficiency
valid_data = df_imputed[(df_imputed['voltage'].notnull()) & 
                       (df_imputed['current'].notnull()) & 
                       (df_imputed['irradiance'].notnull()) & 
                       (df_imputed['irradiance'] > 0)]

if len(valid_data) > 0:
    voltage_irradiance_ratio = (valid_data['voltage'] / valid_data['irradiance']).median()
    current_irradiance_ratio = (valid_data['current'] / valid_data['irradiance']).median()
    
    # Impute voltage using irradiance relationship
    voltage_missing = df_imputed['voltage'].isnull()
    voltage_imputable = voltage_missing & df_imputed['irradiance'].notnull() & (df_imputed['irradiance'] > 0)
    df_imputed.loc[voltage_imputable, 'voltage'] = (
        df_imputed.loc[voltage_imputable, 'irradiance'] * voltage_irradiance_ratio
    )
    print(f"      Imputed {voltage_imputable.sum()} voltage values using irradiance relationship")
    
    # Impute current using irradiance relationship
    current_missing = df_imputed['current'].isnull()
    current_imputable = current_missing & df_imputed['irradiance'].notnull() & (df_imputed['irradiance'] > 0)
    df_imputed.loc[current_imputable, 'current'] = (
        df_imputed.loc[current_imputable, 'irradiance'] * current_irradiance_ratio
    )
    print(f"      Imputed {current_imputable.sum()} current values using irradiance relationship")

print()

# Strategy 2: String-based group imputation (FAST)
print("   Strategy 2: String-based group imputation")

string_impute_cols = ['panel_age', 'maintenance_count', 'soiling_ratio']
for col in string_impute_cols:
    if col in df_imputed.columns:
        before_missing = df_imputed[col].isnull().sum()
        # Group-based imputation
        df_imputed[col] = df_imputed.groupby('string_id')[col].transform(
            lambda x: x.fillna(x.median())
        )
        # If still missing (entire group missing), use global median
        df_imputed[col] = df_imputed[col].fillna(df_imputed[col].median())
        after_missing = df_imputed[col].isnull().sum()
        imputed_count = before_missing - after_missing
        print(f"      {col}: Imputed {imputed_count} values using string_id groups")

print()

# Strategy 3: Simple median imputation for remaining numerical features (VERY FAST)
print("   Strategy 3: Simple median imputation for remaining features")

# Get all numerical columns that still have missing values
numerical_cols = df_imputed.select_dtypes(include=[np.number]).columns.tolist()
numerical_cols = [col for col in numerical_cols if col != 'id']  # Exclude ID

for col in numerical_cols:
    missing_count = df_imputed[col].isnull().sum()
    if missing_count > 0:
        median_value = df_imputed[col].median()
        df_imputed[col] = df_imputed[col].fillna(median_value)
        print(f"      {col}: Imputed {missing_count} values with median ({median_value:.3f})")

print()

print("3. IMPUTATION VALIDATION:")
# Final missing value check
final_missing = df_imputed.isnull().sum()
total_missing = final_missing.sum()

print(f"   Total missing values remaining: {total_missing}")

if total_missing > 0:
    print("   Remaining missing values by column:")
    for col, count in final_missing[final_missing > 0].items():
        percentage = (count / len(df_imputed)) * 100
        print(f"      {col}: {count} ({percentage:.1f}%)")
else:
    print("   SUCCESS: All missing values have been imputed!")

print()

print("4. DATA QUALITY SUMMARY:")
print(f"   Dataset shape: {df_imputed.shape}")
print(f"   Total missing values: {df_imputed.isnull().sum().sum()}")
print(f"   Data types: {len(df_imputed.dtypes.unique())} unique types")
print(f"   Memory usage: {df_imputed.memory_usage(deep=True).sum() / 1024**2:.1f} MB")

print()

print("5. QUICK VALIDATION - SAMPLE STATISTICS:")
key_cols = ['efficiency', 'voltage', 'current', 'irradiance', 'temperature']
for col in key_cols:
    if col in df_imputed.columns:
        print(f"   {col}: mean={df_imputed[col].mean():.3f}, std={df_imputed[col].std():.3f}")

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a copy for outlier handling
df_outlier_handled = df_imputed.copy()

print("1. OUTLIER DETECTION ANALYSIS:")
# Define numerical columns for outlier detection
numerical_cols = ['temperature', 'irradiance', 'humidity', 'panel_age', 'maintenance_count', 
                 'soiling_ratio', 'voltage', 'current', 'module_temperature', 'cloud_coverage', 
                 'wind_speed', 'pressure', 'efficiency']

# Function to detect outliers using IQR method
def detect_outliers_iqr(data, column, multiplier=1.5):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - multiplier * IQR
    upper_bound = Q3 + multiplier * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Analyze outliers in each numerical column
outlier_summary = {}
for col in numerical_cols:
    if col in df_outlier_handled.columns:
        outliers, lower, upper = detect_outliers_iqr(df_outlier_handled, col)
        outlier_count = len(outliers)
        outlier_percentage = (outlier_count / len(df_outlier_handled)) * 100
        
        outlier_summary[col] = {
            'count': outlier_count,
            'percentage': outlier_percentage,
            'lower_bound': lower,
            'upper_bound': upper,
            'min_value': df_outlier_handled[col].min(),
            'max_value': df_outlier_handled[col].max()
        }
        
        print(f"   {col}:")
        print(f"      Outliers: {outlier_count} ({outlier_percentage:.1f}%)")
        print(f"      Range: [{df_outlier_handled[col].min():.3f}, {df_outlier_handled[col].max():.3f}]")
        print(f"      IQR bounds: [{lower:.3f}, {upper:.3f}]")

print()

print("2. STRATEGIC OUTLIER CAPPING:")

# Strategy: Cap outliers using percentile-based approach for RMSE optimization
# Use 1st and 99th percentiles for capping (less aggressive than IQR)

capping_stats = {}
for col in numerical_cols:
    if col in df_outlier_handled.columns:
        # Calculate percentile bounds
        p1 = df_outlier_handled[col].quantile(0.01)
        p99 = df_outlier_handled[col].quantile(0.99)
        
        # Count values that will be capped
        lower_outliers = (df_outlier_handled[col] < p1).sum()
        upper_outliers = (df_outlier_handled[col] > p99).sum()
        total_capped = lower_outliers + upper_outliers
        
        # Apply capping
        original_min = df_outlier_handled[col].min()
        original_max = df_outlier_handled[col].max()
        
        df_outlier_handled[col] = np.clip(df_outlier_handled[col], p1, p99)
        
        capping_stats[col] = {
            'total_capped': total_capped,
            'lower_capped': lower_outliers,
            'upper_capped': upper_outliers,
            'original_range': [original_min, original_max],
            'new_range': [p1, p99]
        }
        
        if total_capped > 0:
            print(f"   {col}: Capped {total_capped} values ({total_capped/len(df_outlier_handled)*100:.1f}%)")
            print(f"      Lower: {lower_outliers} values capped to {p1:.3f}")
            print(f"      Upper: {upper_outliers} values capped to {p99:.3f}")
            print(f"      Range: [{original_min:.3f}, {original_max:.3f}] -> [{p1:.3f}, {p99:.3f}]")
        else:
            print(f"   {col}: No outliers to cap")

print()

print("3. PHYSICS-BASED VALIDATION:")
# Validate that capped values still respect physical constraints
print("   Validating physical constraints after capping:")

# Check efficiency bounds (0 to 1)
efficiency_violations = ((df_outlier_handled['efficiency'] < 0) | 
                        (df_outlier_handled['efficiency'] > 1)).sum()
print(f"      Efficiency outside [0,1]: {efficiency_violations} violations")

# Check soiling_ratio bounds (0.4 to 1.0 based on EDA)
soiling_violations = ((df_outlier_handled['soiling_ratio'] < 0.4) | 
                     (df_outlier_handled['soiling_ratio'] > 1.0)).sum()
print(f"      Soiling_ratio outside [0.4,1.0]: {soiling_violations} violations")

# Check cloud_coverage bounds (0 to 100%)
cloud_violations = ((df_outlier_handled['cloud_coverage'] < 0) | 
                   (df_outlier_handled['cloud_coverage'] > 100)).sum()
print(f"      Cloud_coverage outside [0,100]: {cloud_violations} violations")

# Check for negative irradiance (should be 0 after our earlier fix)
negative_irradiance = (df_outlier_handled['irradiance'] < 0).sum()
print(f"      Negative irradiance: {negative_irradiance} violations")

print()

print("4. OUTLIER IMPACT SUMMARY:")
# Calculate total outliers capped
total_values_capped = sum([stats['total_capped'] for stats in capping_stats.values()])
total_percentage_capped = (total_values_capped / (len(df_outlier_handled) * len(numerical_cols))) * 100

print(f"   Total values capped: {total_values_capped}")
print(f"   Percentage of all numerical values: {total_percentage_capped:.2f}%")

# Show columns with most outliers
most_outliers = sorted(capping_stats.items(), key=lambda x: x[1]['total_capped'], reverse=True)[:5]
print("   Columns with most outliers capped:")
for col, stats in most_outliers:
    if stats['total_capped'] > 0:
        print(f"      {col}: {stats['total_capped']} values")

print()

print("5. DATA QUALITY VALIDATION:")

# Final data quality checks
print(f"   Dataset shape: {df_outlier_handled.shape}")
print(f"   Missing values: {df_outlier_handled.isnull().sum().sum()}")
print(f"   Data types: {len(df_outlier_handled.dtypes.unique())} unique types")

# Key statistics after outlier handling
print("   Key statistics after outlier handling:")
key_cols = ['efficiency', 'voltage', 'current', 'irradiance', 'temperature']
for col in key_cols:
    if col in df_outlier_handled.columns:
        mean_val = df_outlier_handled[col].mean()
        std_val = df_outlier_handled[col].std()
        min_val = df_outlier_handled[col].min()
        max_val = df_outlier_handled[col].max()
        print(f"      {col}: mean={mean_val:.3f}, std={std_val:.3f}, range=[{min_val:.3f}, {max_val:.3f}]")
```

```python
import pandas as pd
import numpy as np
from catboost import CatBoostRegressor
import lightgbm as lgb
from sklearn.model_selection import GroupKFold
from sklearn.linear_model import Ridge, ElasticNet
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb

TRAIN_PATH= "C:\\Users\\samar\\Downloads\\1a513221-3-dataset\\dataset\\zelestra_train.csv"
TEST_PATH= "C:\\Users\\samar\\Downloads\\1a513221-3-dataset\\dataset\\zelestra_test.csv"

# --- LEVEL 1: DIVERSE BASE MODELS ---
def train_level1_models(X, y, X_test, groups):
    """Train diverse Level 1 models with out-of-fold predictions"""
    
    N_SPLITS = 4  # Adjust based on your groups
    gkf = GroupKFold(n_splits=N_SPLITS)
    
    # Out-of-fold predictions for meta-model training
    oof_predictions = np.zeros((len(X), 6))  # 6 base models
    test_predictions = np.zeros((len(X_test), 6))
    
    cat_features = ['string_id', 'error_code', 'installation_type']
    
    for fold, (train_idx, val_idx) in enumerate(gkf.split(X, y, groups)):
        print(f"--- LEVEL 1 FOLD {fold+1}/{N_SPLITS} ---")
        
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
        
        # Model 1: CatBoost Conservative
        model1 = CatBoostRegressor(
            iterations=2500, learning_rate=0.02, depth=8, l2_leaf_reg=3.0,
            random_seed=42, verbose=0, cat_features=cat_features
        )
        model1.fit(X_train, y_train)
        oof_predictions[val_idx, 0] = model1.predict(X_val)
        test_predictions[:, 0] += model1.predict(X_test) / N_SPLITS
        
        # Model 2: CatBoost Aggressive  
        model2 = CatBoostRegressor(
            iterations=2000, learning_rate=0.03, depth=10, l2_leaf_reg=2.5,
            random_seed=123, verbose=0, cat_features=cat_features
        )
        model2.fit(X_train, y_train)
        oof_predictions[val_idx, 1] = model2.predict(X_val)
        test_predictions[:, 1] += model2.predict(X_test) / N_SPLITS
        
        # Model 3: CatBoost Deep
        model3 = CatBoostRegressor(
            iterations=3000, learning_rate=0.015, depth=12, l2_leaf_reg=4.0,
            random_seed=777, verbose=0, cat_features=cat_features
        )
        model3.fit(X_train, y_train)
        oof_predictions[val_idx, 2] = model3.predict(X_val)
        test_predictions[:, 2] += model3.predict(X_test) / N_SPLITS
        
        # Model 4: XGBoost (Algorithm Diversity)
        # Prepare data for XGBoost (no categorical features)
        X_train_xgb = X_train.select_dtypes(include=[np.number])
        X_val_xgb = X_val.select_dtypes(include=[np.number])
        X_test_xgb = X_test.select_dtypes(include=[np.number])
        
        model4 = xgb.XGBRegressor(
            n_estimators=2000, max_depth=6, learning_rate=0.02,
            subsample=0.8, colsample_bytree=0.8, random_state=42
        )
        model4.fit(X_train_xgb, y_train)
        oof_predictions[val_idx, 3] = model4.predict(X_val_xgb)
        test_predictions[:, 3] += model4.predict(X_test_xgb) / N_SPLITS
        
        # Model 5: LightGBM (Algorithm Diversity)
        model5 = lgb.LGBMRegressor(
            n_estimators=2000, max_depth=8, learning_rate=0.02,
            subsample=0.8, colsample_bytree=0.8, random_state=42, verbose=-1
        )
        model5.fit(X_train_xgb, y_train)
        oof_predictions[val_idx, 4] = model5.predict(X_val_xgb)
        test_predictions[:, 4] += model5.predict(X_test_xgb) / N_SPLITS
        
        # Model 6: Random Forest (Algorithm Diversity)
        model6 = RandomForestRegressor(
            n_estimators=1000, max_depth=10, min_samples_split=5,
            random_state=42, n_jobs=-1
        )
        model6.fit(X_train_xgb, y_train)
        oof_predictions[val_idx, 5] = model6.predict(X_val_xgb)
        test_predictions[:, 5] += model6.predict(X_test_xgb) / N_SPLITS
    
    return oof_predictions, test_predictions

# --- LEVEL 2: META-MODEL ---
def train_meta_model(oof_predictions, y, test_predictions):
    """Train meta-model to optimally combine Level 1 predictions"""
    
    print("\n--- TRAINING META-MODEL ---")
    
    # Try multiple meta-models and ensemble them
    meta_models = {
        'ridge': Ridge(alpha=0.1, random_state=42),
        'elastic': ElasticNet(alpha=0.1, l1_ratio=0.5, random_state=42),
        'rf': RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)
    }
    
    meta_predictions = np.zeros(len(test_predictions))
    
    for name, meta_model in meta_models.items():
        print(f"Training {name} meta-model...")
        
        # Train meta-model on out-of-fold predictions
        meta_model.fit(oof_predictions, y)
        
        # Predict on test set
        meta_pred = meta_model.predict(test_predictions)
        meta_predictions += meta_pred / len(meta_models)
        
        # Show feature importance (which base models are most important)
        if hasattr(meta_model, 'coef_'):
            print(f"{name} weights: {meta_model.coef_}")
    
    return meta_predictions

# --- COMPLETE STACKING PIPELINE ---
def run_stacking():
    """Complete stacking implementation"""
    
    # Load and preprocess data (your proven pipeline)
    train_df = pd.read_csv(TRAIN_PATH)
    test_df = pd.read_csv(TEST_PATH)
    
    # Your proven preprocessing
    def robust_clean(df):
        df = df.copy()
        error_strings = ['unknown', 'badval', -999, 'error', 'Error', 'ERROR',
                         'nan', 'NaN', 'NAN', 'null', 'NULL', 'none', 'None', 'NONE']
        df.replace(error_strings, np.nan, inplace=True)
        df.replace(r'(?i)err\d+', np.nan, regex=True, inplace=True)
        
        num_cols = ['temperature', 'irradiance', 'humidity', 'panel_age', 'maintenance_count',
                    'soiling_ratio', 'voltage', 'current', 'module_temperature', 'cloud_coverage',
                    'wind_speed', 'pressure']
        
        for col in num_cols:
            df[col] = pd.to_numeric(df[col], errors='coerce')
            df[col] = df[col].fillna(df[col].median())
            
        df['error_code'] = df['error_code'].fillna('No_Error')
        df['installation_type'] = df['installation_type'].fillna(df['installation_type'].mode()[0])
        df['string_id'] = df['string_id'].fillna('MISSING_STRING')
        return df

    def engineer(df):
        df = df.copy()
        # Your proven features (same as before)
        df['power'] = df['voltage'] * df['current']
        df['temp_differential'] = df['module_temperature'] - df['temperature']
        df['effective_irradiance'] = df['irradiance'] * df['soiling_ratio']
        df['maintenance_frequency'] = df['maintenance_count'] / (df['panel_age'] + 1)
        df['age_soiling_interaction'] = df['panel_age'] * (1 - df['soiling_ratio'])
        df['irradiance_squared'] = df['irradiance'] ** 2
        df['voltage_current_interaction'] = df['voltage'] * df['current']
        df['temp_age_interaction'] = df['module_temperature'] * df['panel_age']
        df['temp_efficiency_loss'] = (df['module_temperature'] - 25) * -0.0045
        df['humidity_temp_stress'] = (df['humidity'] * df['temperature']) / 100
        df['wind_cooling_effect'] = np.log1p(df['wind_speed']) * df['temp_differential']
        df['pressure_efficiency'] = df['pressure'] / 1013.25
        df['thi'] = df['temperature'] * df['humidity'] / 100  
        df['cleaning_effect'] = df['maintenance_frequency'] * (1 - df['soiling_ratio'])
        df['ideal_conditions'] = (df['irradiance'] / 1000) * (1 - df['cloud_coverage']) * (df['wind_speed'] / 10)
        df['panel_degradation'] = 0.005 * df['panel_age'] * (1 - df['maintenance_frequency'])
        df['degradation_env_stress'] = df['panel_degradation'] * df['temp_differential']
        return df

    train_df = engineer(robust_clean(train_df))
    test_df = engineer(robust_clean(test_df))
    
    # Prepare features
    feature_cols = [col for col in train_df.columns if col not in ['id', 'efficiency']]
    X = train_df[feature_cols]
    y = train_df['efficiency']
    X_test = test_df[feature_cols]
    groups = train_df['string_id']
    
    print("ðŸ—ï¸ STARTING STACKING PIPELINE...")
    
    # Level 1: Train diverse base models
    oof_predictions, test_predictions = train_level1_models(X, y, X_test, groups)
    
    # Level 2: Train meta-model
    final_predictions = train_meta_model(oof_predictions, y, test_predictions)
    
    # Minimal post-processing
    final_predictions = final_predictions * 1.005
    max_efficiency = train_df['efficiency'].max() * 1.5
    final_predictions = np.clip(final_predictions, 0, max_efficiency)
    
    # Save submission
    submission = pd.DataFrame({'id': test_df['id'], 'efficiency': final_predictions})
    submission.to_csv("stacking_submission.csv", index=False)
    
    print(f"Prediction range: {final_predictions.min():.4f} - {final_predictions.max():.4f}")
    print(f"Mean prediction: {final_predictions.mean():.6f}")

# Run stacking
run_stacking()
```
